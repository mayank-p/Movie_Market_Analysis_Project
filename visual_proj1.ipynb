Plotted bins of movie budgets and compared them to the profits. Graphed box plots and violin plots

```python
#import objects
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

%matplotlib inline
```

```python
#read data
df = pd.read_csv('data_for_boxplot_original.csv')
```

```python
df.head()
```

```python
#change values of data
df['budget_bin'] = df['budge_bin']/1000000+50
df['profit'] = df['profit']/1000000
```

```python
#create box plot for budget vs profit
sns.set(rc={'axes.facecolor':'#bad4db','figure.facecolor':'#bad4db'})
label_change = pd.DataFrame({'xlabels': ['50-100','100-150','150-200','200-250','250-300','300-350']})
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
plt.ylabel('Profit (million)',size=20)
plt.xlabel('Budget (million)',size=20)
plt.title('Profit at Different Movie Budgets', size = 20)
sns.boxplot('budget_bin', 'profit', data=df)
fig = sns.boxplot('budget_bin', 'profit', data=df)
plt.ylabel('Profit (million)',size=20)
plt.xlabel('Budget (million)',size=20)
plt.tight_layout()
fig.set(xticklabels = ['50-100','100-150','150-200','200-250','250-300','300-350'])
fig.set_xticklabels(['50-100','100-150','150-200','200-250','250-300','300-350'],rotation = 30)
sns.set(rc={'axes.facecolor':'#bad4db','figure.facecolor':'#bad4db'})
plt.savefig("Profit_at_Different_Movie_Budgets.jpg")
```

```python
#create box plot for ratings vs profit
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
sns.boxplot('rating', 'profit', data = df)
plt.xlabel("Average Rating", size=20)
plt.ylabel('Profit (million)',size=20)
plt.title('Profit for Average Ratings',size=20)
```

```python
#create violin plot for budget vs profit
sns.violinplot('budget_bin', 'profit', data=df)
plt.ylabel('Profit (million)',size=20)
plt.xlabel('Budget (million)',size=20)
plt.title('Profit at Different Movie Budgets',size=20)
```

```python
#create violin plot for ratings vs profit
sns.violinplot('rating', 'profit', data=df)
plt.xlabel("Average Rating",size=20)
plt.ylabel('Profit (million)',size=20)
plt.title('Profit for Average Ratings',size=20)
plt.savefig("Profit_at_Different_Movie_Budgets_violin.jpg")
```
